```python
import numpy as np
import sympy as sp
from sympy import Rational

# Question 2.2 npassignment

values = []

a = 5.56
b = -0.5
c = -0.5

l_letters = {0:"s", 1:"p", 2:"d", 3:"f", 4:"g", 5:"h"}

def Eshell(n,l,j):
    return a*(2*n + l - 0.5) + b*(l*(l+1)) + c*(j*(j+1) -l*(l+1) - 3/4)

for n in range(1,2+1):
    for l in range(0,4+1):
        if l == 0:
            j_list = [sp.Rational(1,2)]
        else:
            j_list = [l - sp.Rational(1,2) , l + sp.Rational(1,2)]
        for j in j_list:
            E = Eshell(n, l, j)
            j_frac = sp.nsimplify(j)
            L = l_letters[l] 
            name = f"{n}{L}{j_frac}"
            values.append((E,n,l,j_frac,name))

                          
values_sorted = sorted(values, key=lambda x: x[0])

print("State     n   l    j     E(sorted)")
for E, n, l, j_frac, name in values_sorted:
    print(f"{name:7}  {n:2}  {l:2}  {str(j):4}   {E:.4f}")
```
